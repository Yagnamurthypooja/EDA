```python
def value_to_float(x):
    try:
        if 'K' in x:
            if len(x) > 1:
                return float(x.replace('K', '')) * 0.000001
        elif 'M' in x:
            if len(x) > 1:
                return float(x.replace('M', '')) * 1000000
        elif 'B' in x:
            if len(x) > 1:
                return float(x.replace('B', '')) * 1000000000
        else:
            return float(x)
    except ValueError:
        # Handle the case where the conversion to float fails
        return None

# Apply the function to the '$ Last Change' column
data["$ Last Change"] = data['$ Last Change'].apply(value_to_float)
```
